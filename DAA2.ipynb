```python
def getKurang(myList):
  kurang = 1000
  for row in myList:
      for item in row:
          kurang -= item
  return kurang
getKurang([[100,50,150],[50,250,300]])
```

```python
def searchBinary(myList,item):
  first = 0
  last = len(myList)-1
  foundFlag = False
  while( first<=last and not foundFlag):
    mid = (first + last)//2
    if myList[mid] == item :
        foundFlag = True
    else:
        if item < myList[mid]:
            last = mid - 1
        else:
            first = mid + 1
  return foundFlag
```

```python
searchBinary([8,9,10,100,1000,2000,3000],10)
```

```python
searchBinary([8,9,10,100,1000,2000,3000],5)
```
